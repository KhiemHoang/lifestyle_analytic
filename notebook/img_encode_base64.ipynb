```python
import base64
```

```python
file_path = "/content/drive/MyDrive/Freelance/Cohesion/dashboard_background_dark_theme.png"

with open(file_path, "rb") as f:
    encoded = base64.b64encode(f.read()).decode("utf-8")

data_uri = f"data:image/png;base64,{encoded}"

print(data_uri)
```
